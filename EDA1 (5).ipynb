```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data_clean.csv")
print(data)
```

```python
data.info()
```

```python
print(type(data))
print(data.shape)
print(data.size)
```

```python
data1 = data.drop(['Unnamed: 0', "Temp C"], axis =1)
data1
```

```python
data1.info()
```

```python
data1['Month']=pd.to_numeric(data['Month'],errors='coerce')
data1.info()
```

```python
data1[data1.duplicated(keep = False)]
```

```python
data1.drop_duplicates(keep='first', inplace = True)
data1
```

```python
data1.rename({'Ozones': 'Ozone'}, axis=1, inplace = True)
data1
```

Impute the missing values

```python
data1.info()
```

```python
data1.isnull().sum()
```

```python
cols = data1.columns
colors = ['black','yellow']
sns.heatmap(data1[cols].isnull(), cmap=sns.color_palette(colors), cbar=True)
```

```python
median_ozone = data1["Ozone"].median()
mean_ozone = data1["Ozone"].mean()
print("Median of Ozone: ", median_ozone)
print("Mean of Ozone: ", mean_ozone)
```

```python
data1['Ozone'] = data1['Ozone'].fillna(median_ozone)
data1.isnull().sum()
```

```python
data1['Solar.R'] = data1['Solar.R'].fillna(mean_ozone)
data1.isnull().sum()
```

```python
print(data1["Weather"].value_counts())
mode_weather = data1["Weather"].mode()[0]
print(mode_weather)
```

```python
data1["Weather"] = data1["Weather"].fillna(mode_weather)
data1.isnull().sum()
```

```python
print(data1["Month"].value_counts())
mode_Month = data1["Month"].mode()[0]
print(mode_Month)
```

```python
data1["Month"] = data1["Month"].fillna(mode_Month)
data1.isnull().sum()
```

Detaction of outliers in the columns

Method1: Using histograms and box plots

```python
fig, axes = plt.subplots(2,1,figsize=(8,6),gridspec_kw={'height_ratios':[1,3]})

sns.boxplot(data=data1["Ozone"],ax=axes[0],color='skyblue',width=0.5,orient = 'h')
axes[0].set_title("Boxplot")
axes[0].set_title("Ozone Levels")

sns.histplot(data["Ozone"],kde=True,ax=axes[1],color='purple',bins=30)
axes[1].set_title("Histogram")
axes[1].set_xlabel("Ozone Levels")
axes[1].set_ylabel("Frequency")

plt.tight_layout()

plt.show 
```
